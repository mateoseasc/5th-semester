# Tokenización


### Un token es un caracter
### Un token también puede ser una secuencia de caracteres o palabras, como los ngramas

```python
#3-gramas caracter: "Hol", "a_s", "oy_", ... 
#3-gramas palabras: "Hola soy Mateo", "soy Mateo y", "Mateo y tengo", "y tengo 20", "tengo 20 años"
import nltk
oracion = "Hola soy Mateo y tengo 20 años"
tokens = nltk.word_tokenize(oracion)
print(tokens)

```
